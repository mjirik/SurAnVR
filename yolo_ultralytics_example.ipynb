```python
%matplotlib inline
from pathlib import Path
from random import choice
img_path = choice(list(Path("/auto/plzen1/home/mjirik/data/biomedical/orig/suran_sid/suran_sid/images/test").glob("*.jpg")))
```

```python
from ultralytics import YOLO
from pathlib import Path

model = YOLO("models/detect/train/weights/best.pt")

# predikce na jednom obrázku (uloží výstupní obrázek do runs/detect/predict/)
results = model.predict(
    source=img_path, 
    save=False,
    conf=0.25  # prah pro detekce
)
import matplotlib.pyplot as plt

for r in results:
    plt.imshow(r.plot()[:,:,::-1])  # r.plot() vrátí numpy array s vykreslenými bboxy
    plt.axis("off")
    plt.show()
```
